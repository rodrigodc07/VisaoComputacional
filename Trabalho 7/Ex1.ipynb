```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl
```

```python
def plot_imagem_cinza(img,titulo):    
    plt.imshow(img,cmap = 'gray')
    plt.title(titulo), plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
def plot_imagem_colorido(img,titulo):    
    plt.imshow(img)
    plt.title(titulo), plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
imgname_banda1 = 'LANDSAT_7_ETMXS_20000111_217_076_L2_BAND1.tif'
img_banda1 = cv2.imread(imgname_banda1,0)
plot_imagem_cinza(img_banda1,'Banda 1')
```

```python
imgname_banda2 = 'LANDSAT_7_ETMXS_20000111_217_076_L2_BAND2.tif'
img_banda2 = cv2.imread(imgname_banda2,0)
plot_imagem_cinza(img_banda2,'Banda 2')
```

```python
imgname_banda3 = 'LANDSAT_7_ETMXS_20000111_217_076_L2_BAND3.tif'
img_banda3 = cv2.imread(imgname_banda3,0)
plot_imagem_cinza(img_banda3,'Banda 3')
```

```python
imgname_banda8 = 'LANDSAT_7_ETMPAN_20000111_217_076_L2_BAND8.tif'
img_banda8 = cv2.imread(imgname_banda8,0)
plot_imagem_cinza(img_banda8,'Banda 8')
```

```python
img = cv2.merge((img_banda3,img_banda2,img_banda1))
plot_imagem_colorido(img,'Imagem Colorida')
```
