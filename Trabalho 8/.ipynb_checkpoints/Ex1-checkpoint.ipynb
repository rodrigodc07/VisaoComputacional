# Preâmbulo

A captação de imagens busca, a partir da captura da ondas eletromagneticas, e muitas vezes separamos as ondas entre certos comprimentos de onda por exemplo vermelha 0,61 a 0,68 µm, pois nossos cones já apresentam uma especifidade pare detectar certos coprimentos de onda.

A banda pancromatica é uma unica banda espectral na parte visível do espectro electromagnético, gerando imagens a preto e branco.

# Código

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl
```

```python
def plot_imagem_cinza(img,titulo):    
    plt.imshow(img,cmap = 'gray')
    plt.title(titulo), plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
def plot_imagem_colorido(img,titulo):    
    plt.imshow(img)
    plt.title(titulo), plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
imgname = 'TI_0001.tif'
img = cv2.imread(imgname,1)
plot_imagem_colorido(img,'Bambu')
```
