# Preâmbulo

Operações morfologicas são aquelas que operam sobre a estrutura da imagem binarias, sobre a maneira que os elementos interagem nas diversas vizinhanças.

Onde um elemento estruturante é buscado em cada pixel da imagem, e caso essa configuração seja encontrada o pixel na imagem de saida é ativado, senão ele é desativado

As principais operações morfologicas são: Dilatação, Erosão, Abertura, Fechamento.
