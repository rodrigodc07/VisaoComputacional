# Preâmbulo

Series de fourier são uma maneira discreta de representar elementos continuos como funções, pois eles as transformam em um somatorio de n elementos, e quanto maior o número de n mais preciso é o calculo.

Essa discretização de um elemento continuo em uma amostra de elementos discreto, tambem é bem presente na visão computacional, como por exemplo na transformação de uma imagem continua em uma matriz de intensidade.

# Código

```python
import cv2
import numpy as np
import matplotlib as mpl
import math
from matplotlib import pyplot as plt
from functools import partial
```

```python
def func_triangular(x,repeats):
    somatorio = 0
    for k in range(1,repeats + 1 ):
        razao = (2*k - 1) ** 2
        angulo = (2*k - 1) * math.pi * x
        somatorio = somatorio + (1/razao * np.cos(angulo))
    return 1/2 - (4/(math.pi ** 2)) * somatorio
```

```python
def create_partial_vector_func(function,**kargs):
    func_repets = partial(function,**kargs)
    vector_function = np.vectorize(func_repets)
    return vector_function
```

```python
x = np.linspace(-4,4,100)

for repeats in range(10,110,10):
    vector_func_triangular = create_partial_vector_func(func_triangular,repeats=repeats)
    y = vector_func_triangular(x)
    print("|" + str(repeats) + "|" + str(y.min()) + "|" + str(y.max()) + "|")
    plt.title("Número de senoides: "+ str(repeats))
    plt.plot(x,y)
    plt.xlabel("x")
    plt.ylabel("y") 
    plt.show()
```
